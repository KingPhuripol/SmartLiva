4. ให้น้องเขียนโปรแกรมรับค่า **จำนวนจริงบวก** ไปเรื่อย ๆ ทางหน้าจอจนกว่าจะใส่เลข `-1` และให้หาค่า **น้อยที่สุด** และ **มากที่สุด** ใน List จากนั้นแสดงผลทางหน้าจอ โดยห้ามใช้ฟังก์ชั่น `min` และ `max`

```python
counter = 0
min_value = float('inf')
max_value = float('-inf')
while True:
    number = float(input("Enter a positive number (-1 to stop): "))
    if number == -1:
        break
    if number < min_value:
        min_value = number
    if number > max_value:
        max_value = number
    counter += 1
print("Minimum:", min_value)
print("Maximum:", max_value)
```

1. ดาวินชี่เป็น Data Scientist ที่โรงพยาบาลแห่งหนึ่งที่มีผู้ป่วยไม่มากนัก วันหนึ่งดาวินชี่ได้รับมอบหมายให้สร้างโปรแกรมที่รับค่าอายุของผู้ป่วย แล้วหาค่าเฉลี่ยของอายุผู้ป่วยที่มารับบริการที่โรงพยาบาลที่ดาวินชี่สังกัดอยู่

### Condition:
- จะหยุดรับค่าก็ต่อเมื่อเจอค่าว่าง

### Input
- บรรทัดที่ 1-n: `จำนวนจริง`

### Output
- บรรทัดที่ 1: `ค่าเฉลี่ย`

### Example
| **Input**             | **Output** |
|------------------|--------|
| 25<br>30<br>35<br>40 | 32.5   |
| 50 <br> 60 <br> 70                        | 60.0   |
| 10 <br> 20 <br> 30 <br> 40 <br> 50        | 30.0   |
| 80 <br> 90                                | 85.0   |
| 100                                       | 100.0  |

```python
ages = []

while True:
    try:
        line = input().strip()
        if line == "":
            break
        age = float(line)
        ages.append(age)
    except EOFError:
        break  # กรณีรันในบางระบบที่จบบรรทัดด้วย EOF
    except ValueError:
        continue  # เผื่อผู้ใช้พิมพ์ผิด

# คำนวณและแสดงผล
if len(ages) > 0:
    average = sum(ages) / len(ages)
    print(f"{average:.1f}")
else:
    print("0.0")  # หากไม่มีข้อมูลเลย
```

2. ดาวินชี่ ได้รับมอบหมายอีกครั้ง โดยครั้งนี้ได้รับมอบหมายให้สร้างโปรแกรมที่ตรวจสอบอายุของยา ว่าสามารถใช้งานได้หรือไม่ โดยดาวินชี่จะต้องสร้างฟังก์ชั่นที่รับค่าเป็นอายุของยา (วัน) และความชื้นในอากาศเฉลี่ยในช่วงเวลาที่เก็บรักษายา (%) โดยฟังก์ชี่นจะต้องรีเทิร์นค่ากลับเป็น True/ False

### Condition (Function)
- ถ้ายาอายุเกิน `180 วัน`จะถือว่ายาหมดอายุ
- ถ้าความชื้นเกิน `50%` อายุของยาจะลดลงเหลือ `50%` ของอายุทั้งหมด
- นอกนั้นยายังใช้งานได้ปกติ

Note: ยามีอายุ `180 วัน`

### Condition (หลัก)
- ถ้ายาหมดอายุแล้ว ให้แสดงคำว่า `That mediciene is expired.`
- ถ้ายายังไม่หมดอายุ ให้แสดงคำว่า `That mediciene is safe to take.`

### Input:
- บรรทัดที่ 1: `จำนวนเต็ม` ระบุอายุของยา
- บรรทัดที่ 2: `จำนวนจริง` ระบุความชื้นในกากาศ (%)

### Output:
- บรรทัดที่ 1: คำพูดตาม `Condition`

### Examples:
| **Input** | **Output**                          |
|--------------|------------------------------------------------|
| 150<br>45.0                   | That mediciene is safe to take. |
| 190<br>40.0                   | That mediciene is expired.      |
| 95<br>55.0                   | That mediciene is expired.      |
| 60<br>55.0                    | That mediciene is safe to take. |


```python
def is_medicine_safe(days, humidity):
    MAX_DAYS = 180

    # ถ้าความชื้นเกิน 50% อายุของยาจะลดลงเหลือ 50%
    if humidity > 50.0:
        effective_days = MAX_DAYS * 0.5
    else:
        effective_days = MAX_DAYS

    # ตรวจสอบว่าอายุของยาที่ให้มาเกินอายุที่ยารับได้หรือไม่
    if days > effective_days:
        return False
    else:
        return True

# รับค่าจากผู้ใช้
days = int(input())
humidity = float(input())

# ตรวจสอบผลลัพธ์และแสดงข้อความ
if is_medicine_safe(days, humidity):
    print("That mediciene is safe to take.")
else:
    print("That mediciene is expired.")
```

3. จอห์นเป็นแพทย์แห่งหนึ่งในโรงพยาบาลขนาดใหญ่ที่มีผู้รับบริการวันละหลายคน วันหนึ่งจอห์นสังเกตุว่ามีผู้ป่วยเยอะมาก แต่ชื่อของผู้ป่วยก็ซ้ำกันเยอะมากเช่นเดียวกัน จอห์นนึกสงสัย และอยากรู้ว่ามีผู้ป่วยชื่ออะไรบ้าง จึงได้มอบหมายให้ดาวินชี่ พนักงานตำแหน่ง Data Scientist ให้สร้างโปรแกรมที่รับค่าชื่อของผู้ป่วย แล้วให้แสดงผลชื่อของผู้ป่วยทั้งหมดโดยที่ไม่ให้มีซ้ำกัน

### Condition:
- จะหยุดรับค่าก็ต่อเมื่อเจอค่าว่าง

### Input:
- บรรทัดที่ 1-n: `string` ชื่อของผู้ป่วย

### Output:
- บรรทัดที่ 1-m: `string` ชื่อของผู้ป่วย

### Examples:
| Input                             | Output             |
|-----------------------------------------------|------------------------------|
| Alice <br> Bob <br> Alice <br> Charlie <br> Bob <br> David | Alice <br> Bob <br> Charlie <br> David |
| John <br> Jane <br> Jack <br> John                                                 | John <br> Jane <br> Jack              |
| May <br> May <br> May                                                              | May                                   |
| Anne <br> Ben <br> Cat <br> Dan <br> Elle <br> Ben                                 | Anne <br> Ben <br> Cat <br> Dan <br> Elle |
| Tom <br> Tim <br> Tam <br> Tom <br> Tim                                            | Tom <br> Tim <br> Tam                 |

```python
# เขียนโค้ดที่นี่
def get_unique_names():
    unique_names = []
    seen = set()

    while True:
        name = input().strip()
        if name == "":
            break
        if name not in seen:
            seen.add(name)
            unique_names.append(name)

    # แสดงผล
    for name in unique_names:
        print(name)

# เรียกฟังก์ชัน
get_unique_names()
```

4. ครั้งหนึ่งมีหมู่บ้านแห่งหนึ่งได้ประสบปัญหาโรคระบาดทำให้มีผู้เสียชีวิตเป็นจำนวนมาก เนื่องจากขาดข้อมูลที่สำคัญว่าผู้ป่วยมากจากหมู่บ้านไหนกันแน่เพื่อเตรียมตัวรับมือสถานการณ์ได้อย่างทันท่วงที วันหนึ่งโรงระบาดนี้กลับมาอีกครั้งโดยครั้งนี้ได้ระบาดในหมู่บ้านที่ดาวินชี่ทำงานอยู่ ดาวินชี่จึงได้รับมอบหมายว่าให้เขียนโปรแกรมขึ้นมาเพื่อทำข้อมูล Histrogram ของหมู่บ้านว่าหมู่บ้านไหนที่มีจำนวนผู้ติดเชื้อแล้วมารับการรักษามากที่สุด โดยรับค่าเป็นหมู่บ้าน (เลข) เป็นจำนวนเต็ม แล้วแสดงผลออกมาเป็น Histrogram

### Condition:
- จะหยุดรับค่าเมื่อเจอค่าว่าง

### Input:
- บรรทัดที่ 1 - n: จำนวนเต็ม `1 < x < 999`

### Output:
- บรรทัดที่ 1 - n: Histrogram

### Examples:
| **Input**                             | **Output**                      |
|----------------------------------------|-----------------------------|
| 5<br>1<br>1<br>1<br>1<br>2<br>3<br>4<br>5<br> | [1] ****<br>[2] *<br>[3] *<br>[4] *<br>[5] ** |
| 5<br>2<br>1<br>6 | [1] *<br>[2] *<br>[3]<br>[4]<br>[5] *<br>[6] *|

```python
village_count = {}

while True:
    try:
        line = input().strip()
        if line == "":
            break
        num = int(line)
        if 0 < num < 999:
            village_count[num] = village_count.get(num, 0) + 1
    except:
        break

for key in sorted(village_count):
    stars = '*' * village_count[key]
    print(f"[{key}] {stars}")
```
